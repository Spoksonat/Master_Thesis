```python
sns.set_style("darkgrid")
dark_mode = False
colormap = "Dark2"
black_darkmode = "#28282B"
text_darkmode = "w"
```

```python
def plot(var, signals, signals_name, colors, binss, elements, fills, linestyles, xminv, xmaxv, alphas, linewidths, edgecolors, weightss, n_total, superbin=False, value_superbin=0, log=False, bkg=False,
unity = False):
    
    if(superbin == True):
        new_binss = []
        for n_bin in binss:
            list_bin = list(np.linspace(xminv, value_superbin,n_bin))
            #list_bin.append(xmaxv) make superbin long as the maximum value of the plot
            list_bin.append(10000) # Valor muy largo, tal que cubra seguramente todos los bines
            array_bin = np.array(list_bin)
            new_binss.append(array_bin)
        binss = np.array(new_binss)   

    if(bkg==True):
        plt.figure(figsize=(15,10), facecolor=black_darkmode)

    else:
        plt.figure(figsize=(15,10))

    if(unity == False):
        e = sns.histplot(x=column_names[var], data=df[df["type"] == signals[0]],fill=fills[0], element=elements[0], bins=binss[0], color=colors[0], alpha = alphas[0], linestyle=linestyles[0], 
        linewidth=linewidths[0], ec=edgecolors[0], weights=weightss[0]*np.ones(n_total[0])) # weights=weightss[0]*np.ones(n_total[0])
        for i in range(1,len(signals)):
            f = sns.histplot(x=column_names[var], data=df[df["type"] == signals[i]], fill=fills[i], element=elements[i], bins=binss[i], color=colors[i], alpha=alphas[i], ax=e, linestyle=linestyles[i], 
            linewidth=linewidths[i], ec=edgecolors[i], weights=weightss[i]*np.ones(n_total[i]))

    else:
        e = sns.histplot(x=column_names[var], data=df[df["type"] == signals[0]],fill=fills[0], element=elements[0], bins=binss[0], color=colors[0], alpha = alphas[0], linestyle=linestyles[0], 
        linewidth=linewidths[0], ec=edgecolors[0], stat="density") # weights=weightss[0]*np.ones(n_total[0])
        for i in range(1,len(signals)):
            f = sns.histplot(x=column_names[var], data=df[df["type"] == signals[i]], fill=fills[i], element=elements[i], bins=binss[i], color=colors[i], alpha=alphas[i], ax=e, linestyle=linestyles[i], 
            linewidth=linewidths[i], ec=edgecolors[i], stat="density")

    plt.xlim(xmin=xminv, xmax=xmaxv)

    if(unity == False):
        plt.ylabel("Events", fontsize=20)
    else:
        plt.ylabel("a.u.", fontsize=20)
    plt.xlabel(names[var], fontsize=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    legend = plt.legend(labels=signals_name, fontsize=20)

    if(bkg==True):
        e.tick_params(axis='x', colors=text_darkmode)
        e.tick_params(axis='y', colors=text_darkmode)
        e.xaxis.label.set_color(text_darkmode)
        e.yaxis.label.set_color(text_darkmode)
        legend.get_frame().set_color(black_darkmode)
        for text in legend.get_texts():
            text.set_color(text_darkmode)

        e.set_facecolor(black_darkmode)
        

    if(log==True):
        e.set(yscale="log")
    else:
        e.set(yscale="linear")

    return e   

def scale(histo, binss, scale_factor):
    for i in range(len(binss)):
        if (i == 0):
            for j, patch in enumerate(histo.patches[0:binss[0]]):
                patch.set_height(scale_factor[i]*patch.get_height())
        else:
            for j, patch in enumerate(histo.patches[ np.sum(binss[:i]):np.sum(binss[:i+1]) ]):
                patch.set_height(scale_factor[i]*patch.get_height())

def Number_of_events(signals):
    N = []
    for i,signal in enumerate(signals):
        N.append(df[df["type"] == signal].shape[0]) # Number of rows
    return np.array(N)

def max_height(histo):
    heights = []
    for j, patch in enumerate(histo.patches):
        heights.append(patch.get_height())
    
    heights = np.array(heights)
    return np.amax(heights)

def colormaps(signals, cmap):
    offset = 0
    number_of_colors = 8
    colormap_values = plt.get_cmap(cmap, number_of_colors)
    colormap_values2 = []
    for i in range(number_of_colors):
        colormap_values2.append(colormap_values(i))
    return np.array(colormap_values2)[::-1][offset:] #reversed

def index_kinematic(column_names):
    indexs = []
    for i,name in enumerate(column_names):
        if (("Pt()" in name) or ("M()" in name) or ("Mt()" in name) ):
            indexs.append(i)
    return np.array(indexs)
```

```python
from scipy import interpolate
import matplotlib

min_mass, max_mass = 350, 3000
masses = [350,1000,3000]
n_datos = 100 
xs_bkgs = []
linestyles_bkg = ["-", "-", "-"]

for i,xs in enumerate(cross_sections[3:]):
    xs_bkgs.append([signals_names[3:][i] + " ({0} pb)".format(cross_sections[3:][i]),cross_sections[3:][i]*np.ones(n_datos), linestyles_bkg[i]])

xs_sgns = cross_sections[:3]
func = interpolate.interp1d(masses, np.log10(xs_sgns), kind="quadratic")

new_masses = np.linspace(min_mass, max_mass,n_datos)
new_xs_signs = func(new_masses)

xs_signs = [[r"$Z'$ signals", 10**new_xs_signs, "--"]]

xs_total = np.concatenate((np.array(xs_signs), np.array(xs_bkgs))) # sum of lists = concatenation of lists

labels= [r"$10^{" + str(i) + r"}$" for i in range(-7,2)]

if(dark_mode==True):
    plt.figure(figsize=(15,10), facecolor=black_darkmode)

else:
    plt.figure(figsize=(15,10))

plot = sns.lineplot(x=new_masses, y=xs_total[0][1], label=xs_total[0][0], linewidth=4, linestyle=xs_total[0][2], color=colors[[0,3,4,5]][0])
for i in range(1,len(xs_total)):
    sns.lineplot(x=new_masses, y=xs_total[i][1], label=xs_total[i][0], ax=plot, linewidth=4, linestyle=xs_total[i][2], color=colors[[0,3,4,5]][i])
plot.set(yscale="log")
plot.set_yticklabels(labels,fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel(r"$Z'$ masses (GeV)", fontsize=30)
plt.ylabel(r"Cross section (pb)", fontsize=30)
legend = plt.legend(fontsize=20)

if(dark_mode==True):
    plot.tick_params(axis='x', colors=text_darkmode)
    plot.tick_params(axis='y', colors=text_darkmode)
    plot.xaxis.label.set_color(text_darkmode)
    plot.yaxis.label.set_color(text_darkmode)
    legend.get_frame().set_color(black_darkmode)
    for text in legend.get_texts():
        text.set_color(text_darkmode)
    plot.set_facecolor(black_darkmode)

plt.savefig("Images/Cross_sections.png")

```
